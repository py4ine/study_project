### 형태소 분석

```python
# 라이브러리 불러오기 및 함수화
from konlpy.tag import Komoran
from pykospacing import Spacing
from collections import Counter
import re
import pandas as pd

komoran = Komoran()
spacing = Spacing()
```

```python
# 데이터 읽어옴
def read_data(filename):
    with open(filename, 'r', encoding="cp949") as f:
        data = [line.split('\t') for line in f.read().splitlines()]
        data = data[1:] # 헤더 제거
    return data
```

```python
analysis_data = read_data('./data/contexts_지출.csv')
analysis_data
```

```python
# 데이터 전처리1
def preprocessing(text):
    text = text.lower()  # 소문자 변경
    text = re.sub(r'[^\w\s]', '', text)  # 특수문자 제거
    text = spacing(text)  # 띄어쓰기 보정
    return text
```

```python
preprocessing_data = []

for data in analysis_data:
    data = ''.join(data)
    data = preprocessing(data)
    preprocessing_data.append(data)

preprocessing_data
```

```python
# 형태소 분리
# komoran.morphs(analysis_data)  #형태소
# komoran.nouns(str_to_analyze)  # 명사
# komoran.pos(str_to_analyze)  # 태소를 추출한 뒤 품사 태깅 - 불용어 처리 안됨

# komoran_data = [komoran.morphs(data[1]) for data in preprocessing_data]
komoran_data = []
for data in preprocessing_data:
    morphs = komoran.morphs(data)
    komoran_data.append(morphs)

komoran_data
```

```python
# Word2Vec
from gensim.models import Word2Vec
import time
import numpy as np
```

```python
# Word2Vec의 주요 하이퍼파라미터
# sentences: Word2Vec 모델 학습에 필요한 문장 데이터로 Word2Vec 모델의 입력값으로 사용됩니다.
# size: 단어 임베딩 벡터의 차원(크기)
# window: 주변 단어 윈도우의 크기
# hs: 1(모델학습에 softmax 사용) / 0(negative 옵션 값이 0이 아닐 때 음수 샘플링으로 사용)
    # 음수 샘플링: 학습 과정에서 전체 단어 집합의 임베딩 벡터를 업데이트하지 않고 일부 단어 집합만 업데이트하는 방법 ⇒ 단어 개수가 많아질수록 계산 복잡도가 증가하여 연산 속도가 저하된다는 한계점을 보완
    # 소프트맥스는 빈도가 낮은 단어에 더 잘 작동하고, 음수 샘플링은 빈도가 높은 단어와 저차원 벡터에 더 잘 작동합니다.
# min_count: 단어 최소 빈도 수 제한(설정된 min_count 빈도 수 이하의 단어들은 학습하지 않음)
# sg: 0(CBOW 모델) / 1(skip-gram 모델)

# 학습 시간 측정 시작
start = time.time()
print('Word2Vec 모델 학습 시작')

model = Word2Vec(sentences=token_df[0], vector_size=100, window=4, hs=1, min_count=2, sg=1)
print('Word2Vec 모델 학습 완료 : ', time.time() - start)
```

```python
# 문서 벡터화 함수
def get_document_vector(tokens, model):
    word_vectors = [model.wv[token] for token in tokens if token in model.wv]
    if not word_vectors:
        return np.zeros(model.vector_size)  # 빈 문서에 대한 예외 처리
    return np.mean(word_vectors, axis=0)
```

```python
analysis_data_df = pd.DataFrame(analysis_data, columns=["text"])
```

```python
# 문서 벡터화
analysis_data_df['vector'] = analysis_data_df['text'].apply(lambda x: get_document_vector(x, model))
```
