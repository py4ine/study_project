```python
# 라이브러리 불러오기
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
title = "소비자심리지수"
url = "https://ecos.bok.or.kr/api/StatisticSearch/"
mykey = ""
page = "1/1000"
class_code = "511Y004"  # 대분류코드: 소비자동향조사 2008.9~
time_code = "M"  # 주기(년:A, 반년:S, 분기:Q, 월:M, 반월:SM, 일: D)
start_date = "200401"  # 조회시작날짜
end_date = "202406"  #조회끝날짜
end_class_code = "FME" # 소분류 10101/?/?/?
api_url = url + mykey + "/xml/kr/" + page + "/" + class_code + "/" + time_code + "/" + start_date + "/" + end_date + "/" + end_class_code
api_url
```

```python
response = requests.get(api_url)
response
```

```python
response.content
```

```python
response.text
```

```python
# Data Preprocessing
# web data / API output
soup = BeautifulSoup(response.content, "lxml")
soup
```

```python
data_time_value = []
data_time = soup.find_all('time')
for i in data_time :
    data_time_value.append(i.text)
# data_time_value.reverse()
data_time_value
```

```python
# # 기간이 Quarter 방식일때 실행

# date = []
# for i in data_time_value:
#     year_value = i.split('Q')[0]
#     quarter_value = i.split('Q')[1]
#     if quarter_value == '1':
#         quarter_to_date = "-01-01"
#     elif quarter_value == '2':
#         quarter_to_date = "-04-01"
#     elif quarter_value == '3':
#         quarter_to_date = "-07-01"
#     elif quarter_value == '4':
#         quarter_to_date = "-10-01"
#     date.append(year_value+quarter_to_date)
# print(date)
```

```python
data_value_text = []
data_value = soup.find_all('data_value')
for i in data_value :
    data_value_text.append(i.text)
# data_value_text.reverse()
data_value_text
```

```python
DF_data = pd.DataFrame({"Date": data_time_value, f'{title}': data_value_text})
DF_data['Date'] = pd.to_datetime(DF_data['Date'], format='%Y%m')
DF_data.set_index('Date', inplace=True)
DF_data
```

```python
from datetime import datetime

start_day = '20040101'
end_day = '2024-06-30' # str(datetime.today()-timedelta(1))
start_date = datetime(int(start_day[:4]), int(start_day[4:6]), int(start_day[6:8]))
end_date = datetime(int(end_day[:4]), int(end_day[5:7]), int(end_day[8:10]))

df = pd.DataFrame(index=pd.date_range(start=start_date, end=end_date))
df = df.join(DF_data)
df.fillna(method='ffill', inplace=True)
print(df)
```

```python
df.to_csv(f'./data/api_data_{title}.csv', encoding='cp949')
```
